```python
from typing import Any
from pydantic import BaseModel

class Element(BaseModel):
    type: str
    page_content: Any

# Categorize by type
categorized_elements = []

for element in elements:
    if "unstructured.documents.elements.Table" in str(type(element)):
        categorized_elements.append(Element(type="table",year="2024", page_content=str(element.metadata.text_as_html)))
    elif "unstructured.documents.elements.NarrativeText" in str(type(element)):
        categorized_elements.append(Element(type="text", page_content=str(element)))
    elif "unstructured.documents.elements.ListItem" in str(type(element)):
        categorized_elements.append(Element(type="text", page_content=str(element)))
    elif "unstructured.documents.elements.Title" in str(type(element)):
        categorized_elements.append(Element(type="text", page_content=str(element)))
    elif "unstructured.documents.elements.Address" in str(type(element)):
        categorized_elements.append(Element(type="text", page_content=str(element)))
    elif "unstructured.documents.elements.EmailAddress" in str(type(element)):
        categorized_elements.append(Element(type="text", page_content=str(element)))
    elif "unstructured.documents.elements.Header" in str(type(element)):
        categorized_elements.append(Element(type="CodeSnippet", page_content=str(element)))
    elif "unstructured.documents.elements.CodeSnippet" in str(type(element)):
        categorized_elements.append(Element(type="text", page_content=str(element)))
    elif "unstructured.documents.elements.UncategorizedText" in str(type(element)):
        categorized_elements.append(Element(type="text", page_content=str(element)))
```

```python
def find_longest_text_element(text_elements):
    """Finds the text element with the highest length of page_content and its index."""
    max_length = 0
    longest_text_element = None
    longest_index = -1

    for index, text_element in enumerate(text_elements):  # Use enumerate to get index
        current_length = len(text_element.page_content)
        print(f"Current text length: {current_length}")

        if current_length > max_length:
            max_length = current_length
            longest_text_element = text_element
            longest_index = index  # Store the index of the longest element

    return longest_text_element, max_length, longest_index

# Example usage
longest_element, longest_length, longest_index = find_longest_text_element(text_elements)
if longest_element:
    print(f"Longest text content length: {longest_length}")
    print("Longest text content:", longest_element.page_content)
    print(f"Index of longest text element: {longest_index}")
else:
    print("No text elements found.")
```

```python
# Assuming text_elements is already defined as per your query
concatenated_texts = []
current_concatenation = ""

for text_element in text_elements:
    current_length = len(text_element.page_content)
    print(f"Current text length: {current_length}")

    # Check if the length is less than 300
    if current_length < 350:
        # Concatenate with the previous text
        current_concatenation += text_element.page_content + "/n "  # Add a space for separation
    else:
        # If current_concatenation is not empty, append it to the list
        if current_concatenation:
            concatenated_texts.append(current_concatenation.strip())  # Strip trailing space
            current_concatenation = ""  # Reset for next concatenation
        
        # Append the long text element as is
        concatenated_texts.append(text_element.page_content)

# Append any remaining concatenated text after the loop
if current_concatenation:
    concatenated_texts.append(current_concatenation.strip())

# Output the concatenated texts
print(f"Total concatenated texts: {len(concatenated_texts)}")
for idx, content in enumerate(concatenated_texts):
    print(f"Concatenated Text {idx + 1}: {content}")
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv 

load_dotenv()
```

```python
import os
import time
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate


GROQ_API_KEY = os.getenv("GROQ_API_KEY")

def summaries(text_elements):
    """Summarizes the provided table elements."""
    model = "llama-3.3-70b-versatile"
    summary_chain = (
        {"doc": lambda x: x}
        | ChatPromptTemplate.from_template("Summarize the following html tables or text given to you:\n\n{doc}")
        | ChatGroq(api_key=GROQ_API_KEY, model=model, max_retries=0)
        | StrOutputParser() 
    )
    
    text_content = [i.page_content for i in text_elements]
    
    text_summaries = []
    for i in range(0, len(text_content), 5):  # Process in batches of 5
        batch = text_content[i:i + 5]
        try:
            summaries_batch = summary_chain.batch(batch)
            text_summaries.extend(summaries_batch)
        except Exception as e:
            if "Rate limit reached" in str(e):
                print("Rate limit reached. Waiting to retry...")
                time.sleep(10)  # Wait for 10 seconds before retrying
                continue  # Retry the current batch
            else:
                print(f"An error occurred during summarization: {e}")
                return []
    
    return text_summaries
```

```python
import uuid
from langchain.vectorstores import Chroma
from langchain.storage import InMemoryStore
from langchain.schema.document import Document
from langchain.embeddings import OpenAIEmbeddings
from langchain.retrievers.multi_vector import MultiVectorRetriever
```

```python
# The storage layer for the parent documents
store = InMemoryStore()
id_key = "doc_id"
```

```python
# The vectorstore to use to index the child chunks
vectorstore = Chroma(
    collection_name="financials",
    embedding_function=OpenAIEmbeddings(),
    persist_directory="./chroma_data",
)
```

```python

# The retriever (empty to start)
retriever = MultiVectorRetriever(
    vectorstore=vectorstore,
    docstore=store,
    id_key=id_key,
)
```

```python

# Table summaries embeddings and storage
doc_ids = [str(uuid.uuid4()) for _ in table_elements]
summary_tables = [Document(page_content=s,metadata={id_key: doc_ids[i]}) for i, s in enumerate(table_summaries)]

retriever.vectorstore.add_documents(summary_tables)
retriever.docstore.mset(list(zip(doc_ids, table_elements)))

# Text summaries embeddings and storage
doc_ids = [str(uuid.uuid4()) for _ in text_elements]
summary_texts = [Document(page_content=s,metadata={id_key: doc_ids[i]}) for i, s in enumerate(text_summaries)]

retriever.vectorstore.add_documents(summary_texts)
retriever.docstore.mset(list(zip(doc_ids, text_elements)))
```

```python
from langchain_core.runnables import RunnablePassthrough

# Prompt template
template = """Answer the question based only on the following context, which can include text and tables:
{context}
Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)

# LLM
model = ChatOpenAI(temperature=0, model="gpt-4")

# RAG pipeline
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
```
